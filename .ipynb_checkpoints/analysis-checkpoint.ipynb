```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_name = 'quiz.xlsx'
df = pd.read_excel(file_name)
```

```python
df.head()
```

### Columns, Data type and if Null


```python
df.info()
```

```python

plt.figure(figsize=(8, 6))
sns.barplot(x='Names', y='Score', data=df)

plt.title('Scores by Name')
plt.xlabel('Names')
plt.ylabel('Score')

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Score'], kde=True)
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

## Distribution of answers for each questions

```python
df.columns = df.columns.str.strip()

for column in df.columns[3:]:
    plt.figure(figsize=(4, 2))
    sns.countplot(y=df[column])
    plt.title(f'Question: "{column}"')
    plt.xlabel('Count')
    plt.ylabel('Answers')
    plt.show()
```
